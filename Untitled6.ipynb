```python
# ==============================
# 1. Import libraries
# ==============================
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score, adjusted_rand_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
import json

# ==============================
# 2. Load dataset
# ==============================
df = pd.read_csv("Data Collection for ML mini project (Responses) - Form Responses 1.csv")

print("Dataset shape:", df.shape)
print("\nDataset loaded successfully!")

# ==============================
# 3. Select features
# ==============================
features = [
    'Eating Out Per week',
    'Food Budget per meal Note: Provide integer value for eg like-199',
    'Sweet tooth level (1 is low and 5 is high)',
    'Weekly_hobby_hours   \nprovide values integer between between 0-40'
]

X = df[features].copy()

# Rename for ease
X.columns = ['eat_out', 'food_budget', 'sweet_tooth', 'hobby_hours']

# ==============================
# 4. Data Cleaning
# ==============================
def clean_budget(value):
    if pd.isna(value):
        return np.nan
    value_str = str(value).strip()
    value_str = re.sub(r'[^\d.]', '', value_str)
    try:
        return float(value_str)
    except:
        return np.nan

def clean_hobby_hours(value):
    if pd.isna(value):
        return np.nan
    value_str = str(value).strip()
    value_str = re.sub(r'[^\d.]', '', value_str)
    try:
        return float(value_str)
    except:
        return np.nan

X['eat_out'] = pd.to_numeric(X['eat_out'], errors='coerce')
X['food_budget'] = X['food_budget'].apply(clean_budget)
X['sweet_tooth'] = pd.to_numeric(X['sweet_tooth'], errors='coerce')
X['hobby_hours'] = X['hobby_hours'].apply(clean_hobby_hours)

# Handle missing values
X_cleaned = X.fillna(X.median())

# ==============================
# 5. Outlier Removal (IQR)
# ==============================
def remove_outliers_iqr(df, columns, factor=1.5):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - factor * IQR
        upper_bound = Q3 + factor * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean

X_no_outliers = remove_outliers_iqr(X_cleaned, X_cleaned.columns)
X_final = X_no_outliers.copy()

# ==============================
# 6. Train-Test Split (80/20)
# ==============================
X_train, X_test = train_test_split(X_final, test_size=0.2, random_state=42)

print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# ==============================
# 7. Standardization
# ==============================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ==============================
# 8. Find optimal k using training set
# ==============================
inertia = []
silhouette_scores = []
K = range(2, 11)  # silhouette requires k>=2

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_train_scaled)
    inertia.append(kmeans.inertia_)
    sil_score = silhouette_score(X_train_scaled, kmeans.labels_)
    silhouette_scores.append(sil_score)
    print(f"k={k}: Inertia={kmeans.inertia_:.2f}, Silhouette={sil_score:.3f}")

# Choose best k
best_k = K[np.argmax(silhouette_scores)]
print(f"\nBest k chosen: {best_k}")

# ==============================
# 9. Train final KMeans on training set
# ==============================
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
train_labels = kmeans.fit_predict(X_train_scaled)

# ==============================
# 10. Test the model on 20% test data
# ==============================
test_labels = kmeans.predict(X_test_scaled)

# ==============================
# 11. Evaluation
# ==============================
train_silhouette = silhouette_score(X_train_scaled, train_labels)
test_silhouette = silhouette_score(X_test_scaled, test_labels)

print("\n=== Evaluation Metrics ===")
print(f"Silhouette Score (Train): {train_silhouette:.3f}")
print(f"Silhouette Score (Test):  {test_silhouette:.3f}")

# Adjusted Rand Index between train/test labels (optional extra metric)
# Since clustering is unsupervised, ARI checks consistency
ari = adjusted_rand_score(
    np.concatenate([train_labels, test_labels]),
    np.concatenate([np.repeat("train", len(train_labels)), np.repeat("test", len(test_labels))])
)
print(f"Adjusted Rand Index (Train vs Test consistency): {ari:.3f}")

# ==============================
# 12. Save Results
# ==============================
summary = {
    'Train_Shape': X_train.shape,
    'Test_Shape': X_test.shape,
    'Best_K': best_k,
    'Train_Silhouette': train_silhouette,
    'Test_Silhouette': test_silhouette
}

with open("clustering_eval_summary.json", "w") as f:
    json.dump(summary, f, indent=2)

print("\n✅ Training & Testing complete. Results saved in clustering_eval_summary.json")
```

```python
# ==============================
# 📊 DAY 2 – Exploratory Data Analysis
# ==============================

import matplotlib.pyplot as plt
import seaborn as sns

print("\n=== EDA: Summary Statistics ===")
print(X_final.describe().round(2))

# Histograms for all features
X_final.hist(figsize=(12,8), bins=15, edgecolor='black')
plt.suptitle("Histograms of Features", fontsize=16)
plt.tight_layout()
plt.show()

# Boxplots for outlier inspection
plt.figure(figsize=(12,6))
sns.boxplot(data=X_final)
plt.title("Boxplots of Features", fontsize=14)
plt.show()

# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(X_final.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap", fontsize=14)
plt.show()
```

```python
# ==============================
# 📊 DAY 3 – Cluster Profiling & Visualization
# ==============================

# Refit on the full cleaned data
kmeans_final = KMeans(n_clusters=best_k, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(StandardScaler().fit_transform(X_final))

X_with_clusters = X_final.copy()
X_with_clusters['Cluster'] = cluster_labels

# Cluster centers in original scale
cluster_centers = pd.DataFrame(
    StandardScaler().fit(X_final).inverse_transform(kmeans_final.cluster_centers_),
    columns=X_final.columns,
    index=[f'Cluster {i}' for i in range(best_k)]
)

print("\n=== Cluster Centers (Original Scale) ===")
print(cluster_centers.round(2))

# Cluster profiles (mean values)
for i in range(best_k):
    cluster_data = X_with_clusters[X_with_clusters['Cluster'] == i]
    print(f"\n--- Cluster {i} ---")
    print(cluster_data.describe().loc[['mean','50%']].round(2))  # mean & median

# ---------------- Visualizations ----------------

# Pairplot
sns.pairplot(X_with_clusters, vars=X_final.columns, hue="Cluster", diag_kind="hist", palette="Set2")
plt.suptitle("Pairplot of Features by Cluster", y=1.02, fontsize=16)
plt.show()

# Heatmap of cluster centers
plt.figure(figsize=(10,6))
sns.heatmap(cluster_centers.T, annot=True, cmap="RdYlBu_r", fmt=".2f", linewidths=0.5)
plt.title("Cluster Centers Heatmap")
plt.show()

# Feature distributions per cluster
fig, axes = plt.subplots(2, 2, figsize=(15,12))
axes = axes.ravel()
for i, col in enumerate(X_final.columns):
    for c in range(best_k):
        sns.kdeplot(X_with_clusters[X_with_clusters['Cluster']==c][col], ax=axes[i], label=f'Cluster {c}', fill=True, alpha=0.4)
    axes[i].set_title(f"Distribution of {col}")
    axes[i].legend()
plt.tight_layout()
plt.show()

# Pie chart for cluster sizes
cluster_counts = X_with_clusters['Cluster'].value_counts().sort_index()
plt.figure(figsize=(7,7))
plt.pie(cluster_counts, labels=[f"Cluster {i}" for i in cluster_counts.index], autopct='%1.1f%%', startangle=90)
plt.title("Cluster Size Distribution")
plt.show()

# 3D scatter plot (first 3 features)
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
colors = sns.color_palette("Set2", best_k)
for i in range(best_k):
    subset = X_with_clusters[X_with_clusters['Cluster']==i]
    ax.scatter(subset.iloc[:,0], subset.iloc[:,1], subset.iloc[:,2], s=50, alpha=0.6, color=colors[i], label=f"Cluster {i}")
ax.set_xlabel(X_final.columns[0])
ax.set_ylabel(X_final.columns[1])
ax.set_zlabel(X_final.columns[2])
ax.set_title("3D Cluster Visualization")
ax.legend()
plt.show()
```
